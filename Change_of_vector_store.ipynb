```python

import urllib
import warnings
from pathlib import Path as p
from pprint import pprint

import pandas as pd
from langchain import PromptTemplate
from langchain.chains.question_answering import load_qa_chain
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain.vectorstores import Qdrant
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.embeddings import GooglePalmEmbeddings


warnings.filterwarnings("ignore")
# restart python kernal if issues with langchain import.
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
model = ChatGoogleGenerativeAI(model="gemini-1.5-flash-latest",google_api_key=GOOGLE_API_KEY,
                             temperature=0.2,convert_system_message_to_human=True)
```

```python
pdf_loader = PyPDFLoader("/content/artificial intelligence ☆•☆.pdf")
pages = pdf_loader.load_and_split()
print(pages[3].page_content)
```

```python
embed_model = HuggingFaceEmbeddings(model_name='BAAI/bge-small-en-v1.5')
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
context = "\n\n".join(str(p.page_content) for p in pages)
texts = text_splitter.split_text(context)
```

```python
# prompt: i want to know the leangth of my data split

docs = text_splitter.split_documents(pages)
print(len(docs))

```

```python
from langchain_qdrant import QdrantVectorStore
```

```python
qdrant_url = "QDRANT URL"
qdrant_key = "QDRANT API_KEY"
collection_name = "QDRANT COLLECTION NAME"
```

```python
from langchain.docstore.document import Document

# Initialize QdrantVectorStore with documents and embedding model
# Create Document objects from your texts
documents = [Document(page_content=t) for t in texts]

qdrant = QdrantVectorStore.from_documents(
    documents,             # Pass the list of Document objects
    embed_model,          # Embedding model used to convert documents into vectors
    url=qdrant_url,       # URL for the Qdrant service
    api_key=qdrant_key,   # API key for accessing the Qdrant service
    collection_name=collection_name # Name of the collection to store the vectors in
)
```

```python
vector_index = Chroma.from_texts(texts, embed_model).as_retriever(search_kwargs={"k":5})
```

```python
qa_chain = RetrievalQA.from_chain_type(
    model,
    retriever=vector_index,
    return_source_documents=True

)
```

```python
question = "Describe the Artificall intelligance?"
result = qa_chain({"query": question})
result["result"]
```

```python
question = "Can you tell me the summary of the provided pdf?"
result = qa_chain({"query": question})
result["result"]
```

```python
template = """Use the following pieces of context to answer the question at the end. If you don't know the answer, just say that you don't know, don't try to make up an answer. Keep the answer as concise as possible.Try to make behaviour very politily. Always say "Thank you very much sir Asking question from Atif!" at the end of the answer.
{context}
Question: {question}
Helpful Answer:"""
QA_CHAIN_PROMPT = PromptTemplate.from_template(template)# Run chain
qa_chain = RetrievalQA.from_chain_type(
    model,
    retriever=vector_index,
    return_source_documents=True,
    chain_type_kwargs={"prompt": QA_CHAIN_PROMPT}
)

```

```python
question = "Describe the Artificall intelligance?"
result = qa_chain({"query": question})
result["result"]
```
